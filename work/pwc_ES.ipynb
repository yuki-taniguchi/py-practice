```python
import pandas as pd
```

```python
df = pd.read_csv('dataset/Book1.csv')
df = df.drop(index=0)
```

```python
df.columns
```

```python
df.head(3)
```

```python
df.dtypes
```

```python
for i in df.columns[2:]:
    df[i] = df[i].apply(lambda x: int(x.replace(',', '')))
```

```python
df['県民所得増減'] = df['県民所得_2015'] - df['県民所得_2010']
df['人口増減'] = df['総人口_2015'] - df['総人口_2010']
df['民営事業所増減'] = df['民営事業所数_2015'] - df['民営事業所数_2010']
df['県民所得偏差'] = df['県民所得増減'] - df['県民所得増減'].mean()
```

```python
df['県民所得増減'].mean()
```

```python
df['県民所得増減rank'] = df['県民所得増減'].rank(method='min',ascending=False)
df['人口rank'] = df['総人口_2015'].rank(method='min',ascending=False)
df['人口増減rank'] = df['人口増減'].rank(method='min',ascending=False)
```
