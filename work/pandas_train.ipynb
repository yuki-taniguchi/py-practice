```python
def return_nan_if_zero(x):
    if x == 0:
        return np.nan
    else:
        return x
```

```python
df['budget'] = df['budget'].apply(lambda x: np.nan if x == 0 else x)
df['revenue'] = df['revenue'].apply(lambda x: np.nan if x == 0 else x)
```
